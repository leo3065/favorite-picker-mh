```python
import os
import json
```

```python
imgs_all = os.listdir('../image/official_render')
imgs_rs = os.listdir('../image/rise_sunbreak')
imgs_wi = os.listdir('../image/world_iceborne')
```

```python
imgs_all
```

```python
items = []
item_games = {
    'rise_sunbreak': imgs_rs,
    'world_iceborne': imgs_wi,
}
for img in imgs_all:
    name = img.removesuffix('.png')
    items.append(
        {
            "id": name.lower().replace("'", "-"),
            "name": name.replace("_", " "),
            "image": img,
            "games": list(g for (g, l) in item_games.items() if img in l)
        }
    )
item_data = json.dumps(items)
```

```python
import pyperclip
output = (
    "'use strict';\n"
    "\n"
    f"var items_all = {item_data};\n" 
)
pyperclip.copy(output)
```
